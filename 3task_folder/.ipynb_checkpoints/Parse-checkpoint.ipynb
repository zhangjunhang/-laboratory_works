```python
from bs4 import BeautifulSoup
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
soup = BeautifulSoup(open(r"mine.htm", encoding='utf-8'), "html.parser")
```

```python
# исходные данные, в виде двух таблиц
part1_df = pd.read_html(str(soup.find_all('table')[6].find_all('table')[0]))[0]
part2_df = pd.read_html(str(soup.find_all('table')[6].find_all('table')[1]))[0]

columns = [*['-', 'Сумма'], *part2_df.iloc[0].values]

part1_vals = part1_df[[1,2]].iloc[1:].values.tolist()

for i, row in enumerate(part1_vals):
    part1_vals[i] = [*row, *part2_df.iloc[1:].values[i].tolist()]
all_values = part1_vals

resulting_df = pd.DataFrame(data=all_values, columns=columns)
resulting_df.head(15)
```

```python
resulting_df.to_csv('collected_data.csv', index=False)
```

```python
YAVKA = []
columns = resulting_df.columns
for i in range(1, 33):
    YAVKA.append(round((int(resulting_df.iloc[7, i])+ int(resulting_df.iloc[8, i])) / int(resulting_df.iloc[0, i])*100, 1) )
YAVKA
```

```python
VIKs = []

for i in range(1,31):
    tik_name = r'tik_{}.htm'.format(i)
    soup_tik = BeautifulSoup(open(tik_name, encoding='utf-8'), "html.parser")
    VIKs.append(pd.read_html(str(soup_tik.find_all('table')[6]))[2].shape[1])
VIKs.append(1)
```

```python
plt.figure(figsize=(20,8))
plt.xlim(24, 76)
plt.ylabel('количество участков')
plt.xlabel('явка')

plt.grid()

indexes = np.argsort(YAVKA[1:])

plt.plot(np.array(YAVKA[1:])[indexes], np.array(VIKs)[indexes])
plt.locator_params(axis='x', nbins=20)
plt.title('Зависимость количества участков от явки')
plt.savefig('Зависимость количества участков от явки')
plt.show()
```

Исходя из этих двух графиков(зависимость количества участков и количества избирателей от явки), явно прослеживается линейная зависимость числа участков от количества избирателей. Заметной аномалией является точка при ~53% явки и 9000 количества избирателей. Такой процент явки очень нетипичен.

```python
plt.figure(figsize=(8,8))
plt.ylim(0, 100)
plt.xlim(0, 100)
plt.ylabel('доля голосов')
plt.xlabel('явка')

dolya_1_cand = np.array([float(resulting_df.iloc[12, i].split(' ')[1][:-1]) for i in range(1,33)])
indexes = np.argsort(dolya_1_cand)
plt.plot(dolya_1_cand[indexes], np.array(YAVKA)[indexes])

dolya_2_cand = np.array([float(resulting_df.iloc[13, i].split(' ')[1][:-1]) for i in range(1,33)])
indexes = np.argsort(dolya_2_cand)
plt.plot(dolya_2_cand[indexes], np.array(YAVKA)[indexes])

dolya_3_cand = np.array([float(resulting_df.iloc[14, i].split(' ')[1][:-1]) for i in range(1,33)])
indexes = np.argsort(dolya_3_cand)
plt.plot(dolya_3_cand[indexes], np.array(YAVKA)[indexes])

plt.legend([resulting_df.iloc[12+i, 0] for i in range(3)])
plt.title('Зависимость доли голосов за кандидатов от явки')
plt.savefig('Зависимость доли голосов за кандидатов от явки')
plt.show()
```

Сильной атипичностью является сохранение матожидания при разных кандидатах, но с разницей при разных явках. Это может говорить о том, что в малонаселенных районах за других кандидатов голосовали больше.

```python
import matplotlib.ticker as ticker

fig, ax = plt.subplots(figsize=(8,8))

ax.set_ylim(0.8, 3.2)
ax.set_xlim(-1, 32)

ax.set_ylabel('победитель')
ax.set_xlabel('округ')

TIK_winnner = []
for i in range(2,33):
    TIK_winnner.append(np.argmax( np.array(list(map(lambda x: int(x.split(' ')[0]), resulting_df.iloc[12:15, i].values))) )+1)
ax.scatter(range(1, 32), TIK_winnner)

ax.set_xticks(range(1,32))
ax.set_xticklabels([f'ТИК №{i}' for i in range(1,31)]+[resulting_df.columns[-1]], rotation=90)

ax.set_yticks([1,2,3])
ax.set_yticklabels([resulting_df.iloc[12+i, 0] for i in range(3)])

ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
ax.set_title('Победитель по округу')
plt.savefig('Победитель по округу')
plt.show()
```

Здесь заметна явная аномалия при электронном голосовании. Только в нём победил другой участник.
